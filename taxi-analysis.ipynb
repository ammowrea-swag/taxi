```python
## counting citations at JFK

count = taxi_df['Taxi Pick Up Location'].str.contains('KENNEDY', na=False).sum()
print(f"KENNEDY count: {count}")
```

```python
## counting mentions of LGA

la_guardia_count = taxi_df['Taxi Pick Up Location'].str.contains('LA GUARDIA', na=False).sum()
laguardia_count = taxi_df['Taxi Pick Up Location'].str.contains('LAGUARDIA', na=False).sum()
total = la_guardia_count + laguardia_count
print(f"LA GUARDIA: {la_guardia_count}")
print(f"LaGuardia: {laguardia_count}")
print(f"Total: {total}")
```
